```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('type_of_glass',axis=1)
y=df['type_of_glass']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=35)
```

```python
print("xtrain shape:",xtrain.shape)
print("xtest shape:",xtest.shape)
print("ytrain shape:",ytrain.shape)
print("ytest shape:",ytest.shape)
```

# EVALUATE ALGORITHMS

```python
lg=LogisticRegression()
lg.fit(xtrain,ytrain)
lg.score(xtrain,ytrain)
pred_lg=lg.predict(xtest)
print("Accuracy score of Logistic Regression is:",accuracy_score(ytest,pred_lg))
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
dtc.score(xtrain,ytrain)
dtcpred=dtc.predict(xtest)
print("Accuracy score of DecisionTreeClassifier is:",accuracy_score(ytest,dtcpred))
```

```python
rfc=RandomForestClassifier(n_estimators=100)
rfc.fit(xtrain,ytrain)
pred_rfc=rfc.predict(xtest)
print("Accuracy_score:",accuracy_score(ytest,pred_rfc))
```

```python
gnb=GaussianNB()
gnb.fit(xtrain,ytrain)
pred_gnb=gnb.predict(xtest)
print("Accuracy_score:",accuracy_score(ytest,pred_gnb))
```

```python
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
pred_knn=gnb.predict(xtest)
print("Accuracy_score:",accuracy_score(ytest,pred_knn))
```

```python
svc=SVC()
svc.fit(xtrain,ytrain)
pred_svc=gnb.predict(xtest)
print("Accuracy_score:",accuracy_score(ytest,pred_svc))
```

```python
models=[('Logistic Regression',accuracy_score(ytest,pred_lg)),
       ('KNN',accuracy_score(ytest,pred_knn)),
       ('SVC',accuracy_score(ytest,pred_svc)),
       ('DTC',accuracy_score(ytest,dtcpred)),
       ('GB',accuracy_score(ytest,pred_gnb)),
        ('Randomforest Classifier',accuracy_score(ytest,pred_rfc))
       ]
```

```python
predict=pd.DataFrame(data=models,columns=['models','accuracy'])
predict
```

Observations:
    -The best performance are achieved by Random Forest Classifier.
    -KNN,SVC,GB are badly perform algorithm.
